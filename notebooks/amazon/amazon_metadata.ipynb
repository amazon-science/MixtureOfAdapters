```python
import pandas as pd
import itertools
```

```python
from datasets import load_dataset

dataset = load_dataset("McAuley-Lab/Amazon-Reviews-2023", "raw_meta_Amazon_Fashion", trust_remote_code=True)
```

```python
df = dataset["full"].to_pandas()[['title', 'features', 'description']]
```

```python
texts = list(itertools.chain.from_iterable(list(df['description']))) + list(itertools.chain.from_iterable(list(df['features'])))
```

```python
texts = [text for text in texts if isinstance(text, str) and len(text) > 0]
print(f"Number of texts: {len(texts)}")
```

```python
def extract_sentences_with_keyword(sentences, keywords, must_appear=None):
    keywords = set(k.lower() for k in keywords)  # normalize keywords
    must_appear = set(k.lower() for k in must_appear) if must_appear else None # normalize must_appear
    results = []

    for sentence in sentences:
        sentence = sentence.lower()
        words = set(sentence.split())  # basic tokenization
        if len(words) > 70 or len(words) < 30:
            continue
        
        satisfy = True
        for m in must_appear:
            if m not in sentence:
                satisfy = False

        if not satisfy:
            continue
        
        matched_keywords = [k for k in keywords if k in sentence]

        if len(matched_keywords) == 1:
            results.append((sentence, matched_keywords[0]))

    return results
```

```python
# attribute = 'material'
# pairs = extract_sentences_with_keyword(texts, ['fabric', 'cotton', 'polyester'])

attribute = 'brand'
pairs = extract_sentences_with_keyword(texts, ['nike', 'adidas', 'reebok', 'skechers'])

# attribute = 'weight'
# pairs = extract_sentences_with_keyword(texts, ['3 lb', '5 lb', '16 lb', '22 lb'], must_appear=['food'])
```

```python
df = pd.DataFrame(pairs, columns=['text', attribute])
```
